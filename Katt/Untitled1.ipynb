```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
pathFile = "Resources/cost_of_living_2020.csv"
df = pd.read_csv(pathFile)
df
```

```python
states=["Washington DC", "California", "New York", "Massachusetts", "Washington","Pennsylvania", "Illinois", "Texas", "Georgia"]
# states_df = df[df["State"].str.contains(states)]

states_df = df[df["State"].isin(states)]
statesBar = states_df[["State", "costIndex"]]
statesBar
```

```python
ax = statesBar.plot(kind="bar", x="State", rot=45, title="Cost of Living by State", legend=False, color="purple")
ax.set(ylabel="Cost Index")
plt.tight_layout()
plt.savefig("Resources/Cost_of_Living_by_State.png")
# plt.ylabel("Cost Index")
```
